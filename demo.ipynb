```python
fig, ax = plt.subplots()
fig.axes.append(joyful.lines(90, 60, size=(400,600), title='standard', mode='normal'))
fig.axes.append(joyful.lines(90, 60, size=(400,600), title='bright-peaks', mode='bright'))
fig.axes.append(joyful.lines(90, 60, size=(400,600), title='transparent_mode', mode='transparent'))
```
