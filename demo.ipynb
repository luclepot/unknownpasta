### `joyful.text` (the money shot)

Turns a string into a joydivision cover, with many parameters. 

Play around with configuration etc.


```python
help(joyful.text)
joyful.text("benoit is lame", size=(350,420), spacing=3.0, write_lines=True, weight='light')
```

you can, of course, do it with much longer/more provacative strings.

## `joyful.ascii` 
same as above, but returned in an ascii format rather than an image

```python
asciistr = joyful.ascii("aaron eats poop", 120, 40, hskip=2, remove_whitespace=True)
help(joyful.ascii)
print(asciistr)
```

### `joyful.lines`
We also have the generative function, directly accessible as "lines". These can be really fun too:

**Normal:**

```python
joyful.lines(30,60, title="normal")
```

**Inverted:**

```python
joyful.lines(30, 50, size=(400,500), title='inverted :-0', inverted=True)
```

**a nice gnarly combo:**

```python
joyful.lines(90, 1000, size=(400,600), title='weird', linewidth=0.1, inverted=True, mode='bright')
```
